```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('BRFSS2015.csv')
df.head()
```

```python
df.drop(['AnyHealthcare','NoDocbcCost','MentHlth','PhysHlth','Education','Income','CholCheck'],axis=1,inplace=True)
```

```python
col_names=df.columns
col_names
```

```python
for i in col_names:
    print(df[i].value_counts())
    print("\n\n")
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

# Assuming df is your DataFrame
X = df.drop('HeartDiseaseorAttack', axis=1)
y = df['HeartDiseaseorAttack']

# Undersampling
undersampler = RandomUnderSampler(sampling_strategy=1.0, random_state=42)
X_undersampled, y_undersampled = undersampler.fit_resample(X, y)

# Generating Synthetic Samples using SMOTE
smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# Checking the value counts after undersampling and SMOTE
print("Value counts after undersampling:")
print(y_undersampled.value_counts())

print("\nValue counts after SMOTE:")
print(y_smote.value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
X_train ,X_test,y_train,y_test=train_test_split(X_undersampled,y_undersampled,test_size=0.2)
rfclf=RandomForestClassifier(class_weight='balanced')

rfclf.fit(X_train,y_train)
y_pred=rfclf.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
from joblib import load, dump

dump(rfclf, 'RandomForestClassifier.joblib')
```
